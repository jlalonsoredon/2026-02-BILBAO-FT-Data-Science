## Ejercicio 2
Escribe un programa para pasar de grados a radianes. Hay que usar `input`. Recuerda que la conversión se realiza mediante

$$radianes = grados*(pi/180)$$

````python
import math
PI = math.pi
grados_text = input("Introduce la temperatura en grados Celsius: ")
grados = grados_text.replace(",", ".")  # Reemplaza la coma por un punto si el usuario la usó
try:
    grados = float(grados)
    radianes = grados * (PI / 180)
    print(f"{grados} grados Celsius son {radianes} radianes.")
except ValueError:
    print("Error: Por favor, introduce un número válido para los grados Celsius.")
````

## Ejercicio 3
Escribe un programa que calcule el area de un paralelogramo (base x altura). También con `input`

````python
def calcular_area(base, altura):
    area = (base * altura) / 2
    return area

try:
    base = input("Introduce la base del triángulo: ")
    base = base.replace(",", ".")
    base = float(base)
    
    if base <= 0:
        raise ValueError("La base debe ser un número positivo.")
    
    altura = input("Introduce la altura del triángulo: ")
    altura = altura.replace(",", ".")
    altura = float(altura)
    
    if altura <= 0:
        raise ValueError("La altura debe ser un número positivo.")
    
    area = calcular_area(base, altura)
    print(f"El área del triángulo con base {base} y altura {altura} es: {area}")

except ValueError as e:
    print(f"Error: {e}")
````

## Ejercicio 7
Crea un programa que recoja dos inputs. Tiene que comprobar si su suma es igual, superior o inferior a 10, cada condición por separado

````python
num_a = input("Introduce un número: ")
num_b = input("Introduce otro número: ")

try:
    num_a = float(num_a.replace(",", "."))
    num_b = float(num_b.replace(",", "."))
    
    suma = num_a + num_b
    
    if suma > 10:
        print(f"La suma de {num_a} y {num_b} es mayor que 10.")
    elif suma < 10:
        print(f"La suma de {num_a} y {num_b} es menor que 10.")
    else:
        print("La suma de los números es igual a 10.")
except ValueError:
    print("Error: Por favor, introduce números válidos.")
````

## Ejercicio 8
Razona sin ejecutar código el output que obtendremos de las siguientes sentencias
1. True and True and False

4. not False and True and not True

Compruébalo después con python

1. False
2. not (True) and (True) = False and True = False

4. True and True and False = False

````python
print(True and True and False)                      #Es una contatenación de ANDs por lo que con que haya un False el resultado es FALSE

print(not ((True or False) and (True or False)))    #(True or False) es True, (True or False) es True, entonces (True and True) es True
                                                    #, y el not de True es FALSE
                                                    
print((False or False or False or True) and False)  #(False or False or False Or True)Es una concatenación de ORs 
                                                    #por lo que con que haya un True el resultado es True, 
                                                    #entonces (True and False) es FALSE 
                                                    
print(not False and True and not True)              #not False es True, not True es False, 
                                                    #entonces (True and True and False) al ser uns concatenación de ANDs y 
                                                    # tener un False el resultado es FALSE
````
