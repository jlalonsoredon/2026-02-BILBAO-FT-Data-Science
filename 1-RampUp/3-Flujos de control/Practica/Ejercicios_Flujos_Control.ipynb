## Ejercicio 1
Declara una variable numérica que será una hora cualquiera del día. Implementa la siguiente casuística:

* Si es entre las 0 y las 8, print "Durmiendo"
* Si es entre las 9 y las 18, print "Trabajando"
* Si es entre las 19 y las 21, print "Clase"
* Si es entre las 22 y las 24, print "Descanso"
* En cualquier otro caso, print "Transporte o error"

```python
hora = 24
if 0 < hora < 8:
    estado = "Durmiendo"
elif 8 <= hora < 18:
    estado = "Trabajando"
elif 18 <= hora < 21:
    estado = "Clase"
elif 21 <= hora <= 24:
    estado = "Descansando"
else:
    estado = "Transporte o error"
    
print(f"A las {hora} horas, estoy {estado}")
```

## Ejercicio 2
En este ejercicio vamos a implementar un calculador de precios de casas muy sencillo. Tenemos las siguientes variables:

> superficie
>
> distrito

Implementa la siguiente casuística:

1. Si el distrito es "Moncloa" o "Centro", y además la superficie es superior a 100 metros cuadrados, el precio de la casa es de 1000
2. Si el distrito es "Salamanca", y además la superficie de la casa es al menos de 150 metros, el precio de la casa es de 1500
3. Si el distrito no es "Retiro" y la superficie está entre 60 y 80 metros, el precio es de 600
4. En cualquier otro caso, el precio será de 0

```python
superficie = input("Introduce la superficie de la vivienda en metros cuadrados: ")
try:
    superficie = int(float(superficie))
except ValueError:
    print("Error: La superficie debe ser un número.")
    superficie = 0
distrito = input("Introduce el distrito donde se encuentra la vivienda: ")
try:
    distrito = distrito.capitalize()
except AttributeError:  
    print("Error: El distrito debe ser una cadena de texto.")

if (distrito == "Centro" or distrito == "Moncloa") and superficie > 100:
    precio = 1000
elif (distrito == "Salamanca") and superficie >= 150:
    precio = 1500
elif (distrito != "Retiro") and (superficie >= 60 and superficie <= 80):
    precio = 600
else:
    precio = 0

print(f"El precio de la vivienda es de {precio} euros.")
```

## Ejercicio 8
Implementa un programa que imprima por pantalla el siguiente patrón

5 4 3 2 1 

4 3 2 1 

3 2 1 

2 1 

1

NOTA: NO hay lineas en blanco entre una línea y otra.

```python
num_filas = 5

# realiza una cuenta atras de 5 a 1
for i in range(num_filas, 0, -1):
    #recorre la cuenta atras de 5 a 1 y se le asigna a otro rango que tambien va acia atras de i a 1
    for j in list(range(i, 0, -1)):
        #imprime el rango de j sin salto de linea redefiniendo el parametro end a un espacio 
        print(j, end=" ")
    # al finalizar el bucle de j, se imprime un salto de linea para separar cada cuenta atras más en end 
    # por defecto para hacer un doble salto de linea mas similar a un formato de cuenta atras tradicional a ejemlplo:
    print("\n")
```

## Ejercicio 9
En este ejercicio vamos a crear un pequeño juego. Se trata de intentar adivinar un numero del 1 al 5. Tenemos dos ronda para acertar. Pasos a seguir:

1. Ya viene implementado con la librería `random` cómo obtener un número aleatorio del 1 al 5
2. Tendrás que declarar en una variable el numero de vidas, y mediante un bucle while, comprobar que todavia quedan vidas.
3. Dentro del bucle, obtener el valor del usuario y comprobar si es ese el numero a adivinar. Si no, actualizar las vidas.
4. Si acertamos, salimos del bucle e imprimimos por pantalla "You win". Y si perdemos también salimos del bucle, pero en este caso imprimimos por pantalla "You lose".

TIP: te puede resultar útil usar la sentencia `else` cuando acabe el bucle `while`. Lo que haya dentro de ese `else` se ejecutará una vez acabe la ejecución del `while`. Lo podrás usar para cuando pierdas.

```python
import random
# se genera un número aleatorio entre 1 y 5
num_random = random.randint(1, 5)
# se inicializan los ronda y el máximo de ronda como variables para hacerlo mas confiugrable
ronda = 1
max_ronda = 2

while ronda <= max_ronda:
    # se pide al usuario que adivine el número
    num_usuario = int(input(f"Adivina el número en tu intento {ronda}(entre 1 y 5): "))
    # Si el usuario adivina el número, se muestra un mensaje de felicitación y sale del bucle.
    if num_usuario == num_random:
        print("¡Has ganado!")
        break
    #si el usuario falla pero le quedan ronda, se le informa que vuelva a intentarlo y se incrementa el contador de ronda.
    elif ronda < max_ronda:
        print("Inténtalo de nuevo.")
        ronda += 1
    #si el usuario falla y no le quedan ronda, se muestra un mensaje de que ha perdido y se revela el número correcto.
    else:
        print("HAS PERDIDO.")
        ronda += 1
        print(f"Lo siento, no adivinaste. El número era {num_random}")
```
