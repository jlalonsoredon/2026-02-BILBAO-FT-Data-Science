## Ejercicio Extra
Programa una calculadora que

* Pregunte qué operación quieres realizar (multiplicación/división/suma/resta/potencia/salir)
* Pida dos números 
* Realice la operación
* Siga funcionando hasta que el usuario elija salir

#### Bonus:
* Que guarde todas las operaciones realizadas y las imprima por pantalla cuando el usuario elija esa opción
* Que sea robusta frente a entradas erróneas del usuario o frente a operaciones como dividir por 0

```python
import math
import pprint
import time
```

```python
#inicializamos el historial de operaciones como una lista vacía
historial = []
while True:
    #solicitamos al usuario que introduzca una operación matemática o salir o historial
    operacion = input("Introduce una operación matemática multiplicación * /división / /suma + /resta - /potencia ** /historial/salir")
    #si el usuario introduce salir, se rompe el bucle y se termina el programa
    if operacion.lower() == "salir":
        break
    #si el usuario introduce historial, se recorre el historial de operaciones 
    # y se imprime cada operación con su resultado utilizando eval para calcular el resultado de la operación almacenada en el historial
    # al final continua con la siguiente iteración del bucle
    if operacion.lower() == "historial":
        print("*" * 20, "HISTORIAL", "*" * 20)
        for operacion in historial:
            print(f"{operacion[0]} {operacion[1]} {operacion[2]} = {eval(str(operacion[0]) + operacion[1] + str(operacion[2]))}")
        continue
    # Validamos que opcion introducida sea una operación válida antes de solicitar los números al usuario.
    # si la operación no es reconocida, se muestra un mensaje de error y se continúa con la siguiente iteración.
    if operacion.lower() not in ["multiplicación", "división", "suma", "resta", "potencia", "*", "/", "+", "-", "**"]:
        print("Operación no reconocida")
        continue
    # pide los números al usuario y los convierte float y reemplaza las comas por puntos para evitar errores.
    num1 = float(input("Introduce el primer número").replace(",", "."))
    num2 = float(input("Introduce el segundo número").replace(",", "."))
    # Dependiendo del operador realiza una operación.
    match operacion.lower():
        case "multiplicación" | "*":
            #guardamos la operación en el historial como una tupla con el formato (num1, operador, num2) inmutable
            historial.append((num1,  "*", num2))
            #imprime el resultado de la multiplicación utilizando actual.
            print(f"{num1} * {num2} = {num1 * num2}")
        case "división" | "/":
            #valida la división entre cero.
            if num2 != 0:
                historial.append((num1,  "/", num2))
                print(f"{num1} / {num2} = {num1 / num2}")
            else:
                print("No se puede dividir por cero")
                continue
        case "suma" | "+":
            historial.append((num1,  "+", num2))
            print(f"{num1} + {num2} = {num1 + num2}")
        case "resta" | "-":
            historial.append((num1,  "-", num2))
            print(f"{num1} - {num2} = {num1 - num2}")
        case "potencia " | "**":
            historial.append((num1,  "**", num2))
            print(f"{num1} ** {num2} = {num1 ** num2}")
        case _:
            print("Operación no reconocida")
```
