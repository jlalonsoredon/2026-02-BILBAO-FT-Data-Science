## Ejercicio 1
Dada la siguiente lista:
> ```ejer_1 = [1,2,3,4,5]```

Inviértela para que quede de la siguiente manera
> ```ejer_1 = [5,4,3,2,1]```

````python
ejer_1 = [1,2,3,4,5]
ejer_1.reverse()
print(ejer_1)
````

Otra forma

````python
ejer_1 = [1,2,3,4,5]
new_array = []
i = len(ejer_1)-1
while i >= 0:
    new_array.append(ejer_1[i])
    i = i -1

print(new_array)
````

Otra forma

````python
ejer_1 = [1,2,3,4,5]
new = []
for n in ejer_1[::-1]:
    new.append(n)
print(new)
````

## Ejercicio 3
Crea una lista nueva con todas las combinaciones de las siguientes dos listas:
> ```ejer_3_1 = ["Hola", "amigo"]```
>
> ```ejer_3_2 = ["Que", "tal"]```

Obten el siguiente output:

```['Hola Que', 'Hola tal', 'amigo Que', 'amigo tal']```

````python
ejer_3_1 = ["Hola",  "amigo"]
ejer_3_2 = ["Que", "tal"]
result = []
for i in ejer_3_1:
    for j in ejer_3_2:
        result.append(i + " " +  j)
print(result)
````

## Ejercicio 11
Convierte la siguiente lista en un diccionario
> ```ejer_11 = [("x", 1), ("x", 2), ("x", 3), ("y", 1), ("y", 2), ("z", 1)]```

TIP: Puedes usar dos variables de control en un bucle `for`. Una por cada valor de las tuplas.

````python
ejer_11 = [("x", 1), ("x", 2), ("x", 3), ("y", 1), ("y", 2), ("z", 1)]
my_diccionario = {}
for i in ejer_11:
        my_diccionario[i[0]] = i[1]
print(my_diccionario)
````
