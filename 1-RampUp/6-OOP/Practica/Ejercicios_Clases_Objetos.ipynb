## Ejercicio 1
Somos una empresa que tiene varias tiendas de electrodomésticos. Necesitamos un programa para manejar las tiendas, ver las ventas que han tenido, cuántos empleados hay, etc... Para ello vamos a modelizar la tienda en una clase, que tendrá las siguientes características:
* Nombre clase: "Tienda"
* Atributos comunes:
    * Tipo: "Electrodomésticos"
    * Abierta: True
* Atributos propios de cada tienda:
    * Nombre - String
    * Dirección - String
    * Número de empleados - int
    * Ventas últimos 3 meses - Lista de 3 numeros
* Método para calcular las ventas de todos los meses, que devuelva un número con todas las ventas.
* Método que calcula la media de ventas de los últimos meses, por empleado
* Método que devuelve en un string el nombre de la tienda, junto con su dirección.
* Método que obtiene las ventas del último mes.
* Método para dar una proyección de las ventas en caso de haber invertido X dinero en marketing en esa tienda. Siendo X el parámetro de entrada. Si X es menor de 1000, las ventas de los ultimos 3 meses hubiesen sido de (1.2 x ventas), si es mayor o igual de 1000, las ventas hubiesen sido de (1.5 x venta). El método reescribe el atributo de ventas en los últimos 3 meses, y además devuelve las ventas con el aumento.

Se pide:
1. Implementar la clase "Tienda"
2. Crear tres tiendas con datos inventados
3. Comprueba en al menos una de ellas, todo lo implementado en la clase tienda (sus atributos, media de ventas, ventas/empleado...)
3. Calcular las ventas del último mes de todas las tiendas. Para ello usa el bucle `for`
4. Imprime por pantalla los nombres de las tiendas cuya dirección lleve el string "Avenida"
5. Seguro que ya has hecho este apartado... Documenta la clase :)

```python
class Tienda:
    '''
    Representa una unidad de negocio minorista para la gestión de ventas y empleados.
    
    Esta clase permite administrar la información básica de una tienda, calcular 
    métricas de rendimiento por empleado y proyectar impactos de campañas de marketing.
    '''
    def __init__(self, nombre:str, direccion:str, num_empleados:int, ventas:list, tipo:str = "Electrodomesticos", abierta:bool = True):
        '''
        Inicializa una nueva instancia de Tienda.
        
        Args:
            nombre (str): Nombre comercial de la tienda.
            direccion (str): Ubicación física de la tienda.
            num_empleados (int): Cantidad de personal contratado.
            ventas (list): Lista de valores numéricos que representan las ventas históricas.
            tipo (str): Categoría de productos (por defecto "Electrodomesticos").
            abierta (bool): Estado operativo de la tienda (por defecto True).
        '''
        self.nombre = nombre
        self.direccion = direccion
        self.num_empreados  = num_empleados
        self.ventas = ventas
        self.tipo = tipo
        self.abierta = abierta
        
    def calcular_ventas(self):
        '''
        Calcula el total acumulado de todas las ventas registradas en la lista.
        
        Returns:
            float/int: La suma total de los elementos en self.ventas.
        '''
        return sum(self.ventas)

    def calcular_venta_empleado(self):
        '''
        Calcula el promedio de ventas por cada empleado de la tienda.
        
        Returns:
            float: El resultado de dividir el total de ventas entre el número de empleados.
        '''
        return (self.calcular_ventas() / self.num_empreados)

    def informar_tienda(self):
        '''
        Genera una cadena de texto con los datos identificativos de la tienda.
        
        Returns:
            str: Texto formateado con el nombre y la dirección.
        '''
        informacion = "Nombre: \t\t" + self.nombre + "\nDirección: \t\t" + self.direccion
        return informacion
    
    def ver_ultimas_ventas(self):
        '''
        Obtiene el valor del registro de venta más reciente.
        
        Returns:
            float/int: El último elemento de la lista self.ventas.
        '''
        return self.ventas[-1]
    
    def calcular_retorno_marketing(self, inversion):
        '''
        Aplica un factor de retorno a las ventas actuales basado en una inversión publicitaria.
        
        Modifica la lista self.ventas in-place multiplicando cada registro por un factor
        determinado (1.2 si la inversión es < 1000, 1.5 en caso contrario).
        
        Args:
            inversion (float/int): Cantidad invertida en marketing.
            
        Returns:
            float: El nuevo total de ventas tras aplicar el incremento.
        '''
        retorno = 1.5
        if inversion < 1000:
            retorno = 1.2
        i = 0
        for mes in self.ventas:
            self.ventas[i] = mes * retorno
            i += 1
            
        return (self.calcular_ventas())
```

```python
mi_tienda = Tienda("Electrocasa", "tabira 20", 3, [1000, 2000, 5000])
tienda_bilbao = Tienda("ElectroBilbao", "gran via 1", 5, [10000, 20000, 30000])
tienda_barakaldo = Tienda("tiendaBaraka", "Avenida Nagusia 3", 2, [100, 200, 400])


print(f"Datos de :\n\
{mi_tienda.informar_tienda()}\n\
Ventas de los 3 meses:  {mi_tienda.calcular_ventas()}\n\
Media por empleado ({mi_tienda.num_empreados}): {mi_tienda.calcular_venta_empleado()}\n\
Ventas último mes:      {mi_tienda.ver_ultimas_ventas()}\n\
Proyección de Marketing: {mi_tienda.calcular_retorno_marketing(1600)}")
print("Lista de ventas actulizada con la proyección de Marketing: ",*mi_tienda.ventas)

todas_tiendas = [mi_tienda, tienda_bilbao, tienda_barakaldo]

for tienda in todas_tiendas:
    total_ventas += tienda.calcular_ventas() 

print("Ventas totales: ", total_ventas)

busqueda = "Avenida"
for tienda in todas_tiendas:
    if busqueda.lower() in tienda.direccion.lower():
        print("Tienda encontrada con :", busqueda)
        print(tienda.informar_tienda())
```

## Ejercicio 2
En este ejercicio vamos a implementar una clase *Perro* en Python. La clase tiene las siguientes características:
* Cosas que sabemos seguro que tiene un perro
    * Tiene 4 patas
    * 2 orejas
    * 2 ojos
    * Una velocidad de 0. Por defecto, el perro está parado
* Cuando se inicialice:
    * El perro será de una determinada raza
    * Por defecto tendrá pelo "Marrón", a no ser que se diga lo contrario.
    * Por defecto no tendrá dueño, a no ser que se diga lo contrario.
    
* Dispondrá también de un método llamado andar, que tiene un argumento de entrada (aumento_velocidad). Este valor se le sumará a la velocidad que ya llevaba el perro.
* Necesita otro método (parar), donde pondremos la velocidad a 0.
* Otro método llamado "ladrar", que tendrá un argumento de entrada, y la salida será el siguiente string: "GUAU!" + el argumento de entrada.


Se pide:
* Implementa la clase *Perro*
* Crea un objeto de tipo *Perro*, sin dueño
* Comprueba que están bien todos sus atributos
* Prueba que ande, y comprueba su velocidad
* Páralo
* Documenta la clase *Perro*

```python
class Perro:
    def __init__(self, raza:str, pelo:str = "marron", dueno:bool = False , patas:int = 4, orejas:int = 2, ojos:int = 2, velocidad:float = 0):
        self.raza = raza
        self.pelo = pelo
        self.dueno = dueno
        self.patas =patas
        self.orejas = orejas
        self.ojos = ojos
        self.velocidad = velocidad
        
    def andar(self, aumento_velocidad):
        self.velocidad += aumento_velocidad
    
    def parar(self):
        self.velocidad = 0
    
    def ladrar(self, texto):
        ladrido = "GUAU!" + " " + texto
        return ladrido
```

```python
rufus = Perro("pastor", "verde", False)
print(rufus.raza)
print(rufus.pelo)
print(rufus.dueno)
print(rufus.patas)
print(rufus.orejas)
print(rufus.ojos)
print(rufus.velocidad)

print(rufus.__dict__)

rufus.andar(10)
print(f"la velocidad actual es de: {rufus.velocidad} km/h")
rufus.parar()
print(f"la velocidad actual es de: {rufus.velocidad} km/h")

print(rufus.ladrar("Che"))
```

```python
class Perro:

    '''
    Clase para modelar los perros de mi perrera

    Attributes:
        patas (int): númera de patas
        orejas (int): número de orejas
        ojos (int): número de ojos
        velocidad (float): velocidad en Km/h a la que va el perro
    '''
    
    patas = 4
    orejas = 2
    ojos = 2
    velocidad = 0

    def __init__(self, raza, pelo="Marrón", dueño=False):
        '''
        Args:
            raza (str): Raza del perro
            pelo (str): Color del pelo
            dueño (bool): Indicador de si tiene dueño
        '''
        self.raza = raza
        self.pelo = pelo
        self.dueño = dueño

    def andar(self, aumento_v):
        '''
        Hace andar al perro añadiendole una velocidad
        
        Args:
            aumento_v (float): aumento de la velocidad del perro
        Returns:
            str con la velocidad a la que va el perro
        '''
        self.velocidad = self.velocidad + aumento_v
        return "Perro andando con velocidad de " + str(self.velocidad)
    
    def parar(self):
        '''
        Hace para al perro con una velocidad por tanto de 0

        Returns:
            str con la velocidad a la que va el perro de 0
        '''
        self.velocidad = 0
        return "Perro parado con velocidad de " + str(self.velocidad)

    def ladrar(self, msg):
        '''
        El perro ladra, con un mensaje

        Args:
            msg (str): mensaje que ladra el perro
        Returns:
            str con un ladrido y el mensaje
        '''
        return "GUAU! " + msg
```
