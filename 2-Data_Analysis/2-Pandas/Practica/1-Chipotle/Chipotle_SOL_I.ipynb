### Step 1. Import the necessary libraries

```python
import pandas as pd
import numpy as np
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv) and assign it to a variable called chipo.

```python
url = "https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv"
chipo = pd.read_csv(url, sep="\t")
chipo
```

### Step 3. Name of the max valued product

```python
chipo["item_price"] = chipo["item_price"].str.replace('$', '').astype(float)
#creamos una columna con el precio unitario del producto
chipo["unit_price"] = chipo["item_price"] / chipo["quantity"]
```

```python
chipo
```

```python
chipo[chipo["item_price"] == chipo["item_price"].max()]["item_name"].values[0]
```

### Step 4. How many products cost more than $10.00?

```python
chipo[chipo["unit_price"] > 10.00]["item_name"].nunique()
```

### Step 4.1: Y cuántos pedidos se han hecho con un producto de más de 10$? Es lo mismo?

```python
chipo[chipo["unit_price"] > 10.00]
```

### Step 4.2: Y cuántos pedidos se han hecho de más de 10$? Es lo mismo?

```python
chipo[(chipo["item_price"] > 10.00)]["quantity"].sum() 
```

### Step 4.3: Y en cuántos pedidos se ha pagado más de 10$ por un mismo producto? Es lo mismo?

```python
print(chipo[chipo["item_price"] > 10.00]["order_id"].nunique())
```

### Step 5. What is the price of each item and name it unit_price. Get only item_name and unit_price

```python
tabla_precios = chipo[["quantity", "item_name", "item_price"]].copy()
tabla_precios["unit_price"] = tabla_precios["item_price"] / tabla_precios["quantity"]
tabla_precios = tabla_precios[["quantity","item_name", "item_price", "unit_price"]].drop_duplicates()
tabla_precios
```

### Step 6. Sort by the name of the item

```python
tabla_precios.sort_values(by = "item_name")
```

### Step 7. What was the quantity of the most expensive item ordered? 2 ways

```python
print(f'Cantidad de productos: {chipo.loc[chipo["item_price"].idxmax(), "quantity"]}')

chipo.loc[chipo["item_price"].idxmax()]
```

```python
chipo.sort_values(by = "item_price", ascending = False).head()
```

### Step 8. How many times was a Veggie Salad Bowl ordered?

```python
lineas_veggie_salad = chipo[chipo["item_name"] == "Veggie Salad Bowl"]

print(lineas_veggie_salad["quantity"].sum())
lineas_veggie_salad
```
