9. What is the name of the oldest person who died in the Titanic? Was he or she travelling alone or had any family travelling with them?  



In order to give an answer to the second question you should find out which columns give you that info. Usually part of your job as a Data Scientist will be get to know the dataset which you are working with. In this case the columns which give you that info are the following: 
        - 'sibsp' Number of Siblings/Spouses Aboard
        - 'parch' Number of Parents/Children Aboard

```python
max_age = titanic['age'].max()
#print(max_age)

olders = titanic[["survived", "name", "age", "sex", "sibsp", "parch"]][(titanic["age"] == max_age) & (titanic["survived"] == 0)]
print("murieron:", len(olders), "personas de", max_age, "años")
olders
titanic[titanic["age"] == max_age]
```

```python
deaths = titanic[titanic["survived"] == 0]
max_age_deaths = deaths["age"].max()
max_age_deaths
older_deaths = deaths[deaths["age"] == max_age_deaths]
print("murieron:", len(older_deaths), "personas de", max_age_deaths, "años")

for index, row in older_deaths.iterrows():
    print("El pasajero", row['name'], "de", row['age'], "años, era del sexo", row['sex'])
    if row['sibsp'] > 0 & row['parch'] > 0:
        print("Viajaba con familiares")
    else:
        print("Viajaba solo")
older_deaths
```

19. Drop columns in `titanic` with more than 1000 missing values and print the columns remaining

```python
# Contar valores faltantes en cada columna
missing_count = titanic.isnull().sum()
print("Valores faltantes por columna:")
print(missing_count)

# Identificar columnas con más de 1000 valores faltantes
cols_to_drop = missing_count[missing_count > 1000].index.tolist()
print("\nColumnas a eliminar (más de 1000 NaN):", cols_to_drop)

# Eliminar esas columnas
titanic_cleaned = titanic.drop(columns=cols_to_drop)

# Mostrar columnas restantes
print("\nColumnas restantes:", titanic_cleaned.columns.tolist())
```

20. Calculate the ratio of missing values at the `boat` column. 

```python
# Calcular el número de valores faltantes en la columna 'boat'
missing_boat = titanic['boat'].isnull().sum()

# Calcular el total de filas
total_rows = len(titanic)

# Calcular el ratio
ratio_missing = missing_boat / total_rows

print(f"Valores nulos en 'boat': {missing_boat}")
print(f"Total de filas: {total_rows}")
print(f"Ratio de valores nulos: {ratio_missing:.4f} ({ratio_missing * 100:.2f}%)")
```

24. Using [`.pivot_table()`](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.pivot_table.html) to count how many women or men survived by class, call it `counted`.

Don't panic and read the documentation!




```python
counted = titanic.pivot_table(
    index='pclass', 
    columns='gender', 
    values='survived', 
    aggfunc='sum'
)
counted 
```

25. Add a new column with the sum of survived men and women, call it `counted['total']` 


```python
counted['total'] = counted.sum(axis=1)
counted
```

26. Sort `counted` by the `'total'` column. In which class the people survived the most?

30. Find if was there any `name` repeated at the Titanic?

Hint: There were two people with the same name, who?

```python
# Encontrar nombres que aparecen más de una vez
duplicated_names = titanic["name"].value_counts()[titanic["name"].value_counts() > 1]
print("Nombres repetidos:")
print(duplicated_names)

# Mostrar las filas con nombres duplicados
repeated_passengers = titanic[titanic["name"].isin(duplicated_names.index)]
print("\nPasajeros con nombres duplicados:")
repeated_passengers[["name", "age", "gender"]]
```

33. Using the function [`catplot`](https://seaborn.pydata.org/generated/seaborn.catplot.html) from the library `seaborn`, find out if the hypothesis _"Women are more likely to survive shipwrecks"_ is true or not.

You should get something like this:

```python
sns.catplot(x='gender', y='survived', data=titanic, kind='bar', hue='gender', legend=True)
plt.title('Tasa de Supervivencia por Género')
plt.ylabel('Tasa de Supervivencia')
plt.xlabel('Género')
plt.show()

# Análisis: Si la barra de "female" es más alta (cercana a 1), la hipótesis es verdadera
```

34. Using [`kdeplot`]("https://seaborn.pydata.org/generated/seaborn.kdeplot.html") from `seaborn` represent those who not survived distributed by age.<>

Hint: First you should "filter" the `titanic` dataset where the column "survived" is 0, indexing the column `"age"` only.

Arguments you should pass to the function: 

    - color = "red"
    - label = "Not Survived"
    - shade = True
    
You should get something like this: 

![kdeplot](./img/kdeplotsur.png)

```python
# Filtrar pasajeros que NO sobrevivieron (survived == 0)
not_survived = titanic[titanic['survived'] == 0]

# Crear el gráfico KDE
plt.figure(figsize=(10, 6))
sns.kdeplot(data=not_survived, x='age', fill=True, color='red', label='Not Survived')
plt.title('Distribución de Edad de los Pasajeros que NO Sobrevivieron')
plt.xlabel('Edad')
plt.ylabel('Densidad')
plt.legend(["Not Survived"])
plt.show()
```
